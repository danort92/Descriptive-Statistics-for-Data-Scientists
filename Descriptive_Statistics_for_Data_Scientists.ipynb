```r
dati=read.csv("https://raw.githubusercontent.com/danort92/Profession.AI/main/realestate_texas.csv",sep=",")



#CALCULATE POSITION, VARIABIALITY, FORM INDICES AND FREQUENCY DISTRIBUTIONS FOR THE VARIABLES

#packages for form indices
install.packages("moments")
library(moments)

attach(dati)
n=dim(dati)[1]

#creates function for variability index (not built-in function)
cv=function(x){
  sd(x)/mean(x)*100
}

#creates function for Gini index (not built-in function)
gini.index=function(x){
  ni=table(x)
  fi=ni/length(x)
  fi2=fi^2
  J=length(table(x))
  gini=1-sum(fi2)
  gini.norm=gini*J/(J-1)
  return(gini.norm)
}

#creates function for range index (not built-in function)
range=function(x){
  return(max(x)-min(x))
}



#1-city: qualitative variable - let's just calculate frequency distribution
ni=table(city)
fi=ni/n
Ni=cumsum(ni)
Fi=Ni/n
cbind(ni,fi,Ni,Fi)

#2-year: it's useful for line charts - let's just calculate frequency distribution
ni=table(year)
fi=ni/n
Ni=cumsum(ni)
Fi=Ni/n
cbind(ni,fi,Ni,Fi)

#3-month: it's useful for line charts - let's just calculate frequency distribution
ni=table(month)
fi=ni/n
Ni=cumsum(ni)
Fi=Ni/n
cbind(ni,fi,Ni,Fi)
#divide in trimesters, just to avoid having 12 initial modalities
dati$quarter=cut(month, breaks=c(0,3,6,9,12))
attach(dati)
ni=table(quarter)
fi=ni/n
Ni=cumsum(ni)
Fi=Ni/n
cbind(ni,fi,Ni,Fi)

#4-sales: quantitative variable - let's calculate indices
min(sales)
max(sales)
mean(sales)
median(sales)
quantile(sales)
summary(sales)

range(sales)
IQR(sales)
var(sales)
sd(sales)
cv(sales)
gini.index(sales)

skewness(sales)
kurtosis(sales)-3

#5- volume: quantitative variable - let's calculate indices
min(volume)
max(volume)
mean(volume)
median(volume)
quantile(volume)
summary(volume)

range(volume)
IQR(volume)
var(volume)
sd(volume)
cv(volume)
gini.index(volume)

skewness(volume)
kurtosis(volume)-3

#6-median_price: quantitative variable - let's calculate indices
min(median_price)
max(median_price)
mean(median_price)
median(median_price)
quantile(median_price)
summary(median_price)

range(median_price)
IQR(median_price)
var(median_price)
sd(median_price)
cv(median_price)
gini.index(median_price)

skewness(median_price)
kurtosis(median_price)-3

#7-listings: quantitative variable - let's calculate indices 
min(listings)
max(listings)
mean(listings)
median(listings)
quantile(listings)
summary(listings)

range(listings)
IQR(listings)
var(listings)
sd(listings)
cv(listings)
gini.index(listings)

skewness(listings)
kurtosis(listings)-3

#8-months_inventory: quantitative variable - let's calculate indices
min(months_inventory)
max(months_inventory)
mean(months_inventory)
median(months_inventory)
quantile(months_inventory)
summary(months_inventory)

range(months_inventory)
IQR(months_inventory)
var(months_inventory)
sd(months_inventory)
cv(months_inventory)
gini.index(months_inventory)

skewness(months_inventory)
kurtosis(months_inventory)-3

#It can be observed that:
#the variables with the highest and the lowest variability are respectively volume and median_price
#the variables with the highest and the lowest symmetry are respectively volume and months_inventory



#plots volume to verify its positive asymmetry (higher frequency of low values) 
install.packages("ggplot2")
library(ggplot2)
ggplot()+
  geom_density(aes(x=volume),col="black",
               fill="lightblue")+
  labs(title="Probability density for variable volume",
       x="Total price in mln$",
       y="Relative frequency")+
  geom_vline(xintercept = mean(volume),col="red",lwd=1)+
  geom_vline(xintercept = median(volume),col="blue",lwd=1)+
  geom_label(aes(y=0.002,x=mean(volume)+8),
            label="mean", col="red")+
  geom_label(aes(y=0.002,x=mean(volume)-13),
             label="median", col="blue")



#comparison between normalized variables volume and median_price
#volume is more asymmetric than median_price and noticeably has a higher kurtosis
ggplot()+
  geom_density(aes(x=(volume-mean(volume))/sd(volume)),col="blue")+
  geom_density(aes(x=(median_price-mean(median_price))/sd(median_price)),col="red")+
  labs(title="Comparison between probability density of normalized variables volume and median_price",
       x="Normalized values",
       y="Relative frequency")



#let's take a quantitative variable, divide it in classes and show its frequency distribution, bar plot and let's calculate its Gini index

dati$sales_cl=cut(sales,breaks=seq(75,425,70))
attach(dati)
ni=table(sales_cl)
fi=ni/n
Ni=cumsum(ni)
Fi=Ni/n
cbind(ni,fi,Ni,Fi)

barplot(ni,
        main="Frequency distribution for variable sales in classes",
        xlab="N° of sales in classes",
        ylab="Absolute frequency",
        col="pink",
        border="red")
barplot(fi,
        main="Frequency distribution for variable sales in classes",
        xlab="N° of sales in classes",
        ylab="Relative frequency",
        col="lightblue",
        border="blue")

gini.index(sales_cl)



#Let's calculate the probability that, casually choosing a row of this dataset, it containes 
#1) Beaumont in the variable city, 2) July in the variable month and 3) December in the variable month and at the same time 2012 in the variable year

table(city)["Beaumont"]/n
table(month)["7"]/n
table(month)["12"]/n*table(year)["2012"]/n



#Let's create a new dataset variable: the mean_price
dati$mean_price=volume*10^6/sales



#Let's try to create a variable that gives the idea of how effective are the sales listings
dati$effectiveness=(sales/listings)*100/months_inventory



#let's summarise the indices relating them to the varibales city, year and/or month

install.packages("dplyr")
library(dplyr)

#evaluates mean,std,median and variability coefficients as a function of city
dati %>%
  group_by(city)%>%
  summarise(media=mean(sales),
            sd=sd(sales),
            mediana=median(sales),
            coeff.var=cv(sales),
  )


#evaluates mean,std,median and variability coefficients as a function of year
dati %>%
  group_by(year)%>%
  summarise(media=mean(sales),
            sd=sd(sales),
            mediana=median(sales),
            coeff.var=cv(sales),
  )

#evaluates mean,std,median and variability coefficients as a function of month trimesters
dati %>%
  group_by(quarter)%>%
  summarise(media=mean(sales),
            sd=sd(sales),
            mediana=median(sales),
            coeff.var=cv(sales),
  )

#evaluates mean,std,median and variability coefficients as a function of city for each year
dati %>%
  group_by(city,year)%>%
  summarise(media=mean(sales),
            sd=sd(sales),
            mediana=median(sales),
            coeff.var=cv(sales),
  )



#Let's evaluate through boxplots the distribution of median_price as a function of city and year
ggplot(dati)+
  geom_boxplot(aes(x=city,y=median_price))+
  labs(title="Boxplot of median price per city",
       x="City",
       y="Median selling price [$]")

ggplot(dati)+
  geom_boxplot(aes(x=city,y=median_price,fill=factor(year)))+
  labs(title="Boxplot of median price per city each year",
       x="City",
       y="Median selling price [$]",
       fill="Anni")

#Let's use gghalves to show violin plot and have a look both to variaiblity and form indices
install.packages("gghalves")
library(gghalves)
ggplot(dati)+
  geom_half_boxplot(aes(x=city,y=median_price),side="l",fill="lightblue")+
  geom_half_violin(aes(x=city,y=median_price),side="r",fill="pink")+
  labs(title="Boxplot and probability density of median price per city",
       x="City",
       y="Median selling price [$]")

#summarises some important indices all together
dati %>%
  group_by(city)%>%
  summarise(media=mean(median_price),
            mediana=median(median_price),
            IQR=IQR(median_price),
            coeff.var.=cv(median_price),
            asimmetria=skewness(median_price),
            curtosi=kurtosis(median_price)-3,
  )



#Let's use boxplot for the variable sales too
ggplot(dati)+
  geom_boxplot(aes(x=city,y=sales,fill=factor(year)))+
  labs(title="Boxplot of n° of sales per city each year",
       x="City",
       y="N° of sales",
       fill="Years")



#Let's use a bit of ggplot to show some variables as a function of city, year and month

#if geom_bar is used stat=identity must be used, if geom_col is used it's not necessary
ggplot(dati)+ 
  geom_bar(aes(x=city,y=sales, fill=year), stat="identity", position="stack")+
  labs(title="Bar plot of n° of sales each year as a function of city",
       x="City",
       y="N° of sales",
       fill="Years")

ggplot(dati)+ 
  geom_col(aes(x=city,y=sales, fill=year), position="stack")+
  labs(title="Bar plot of n° of sales each year as a function of city",
       x="City",
       y="N° of sales",
       fill="Years")

#normalized plot (position=fill)
ggplot(dati)+ 
  geom_bar(aes(x=city,y=sales, fill=year), stat="identity", position="fill")+
  labs(title="Bar plot of n° of sales each year as a function of city",
       x="City",
       y="N° of sales",
       fill="Years")



#Let's create a line chart of sales as a function of city, year and month

#generates array containing the n° sales per year and than plots the result
sales_history_year=c()
sales_history_month=c()
k=1
for (i in min(year):max(year)){
  for (j in min(month):max(month)){
  sales_history_month[j]=sum(sales[year==i & month==j])
  }
  sales_history_year[k]=sum(sales_history_month)
  k=k+1
}

ggplot()+
  geom_line(aes(x=c(min(year):max(year)),y=sales_history_year),lwd=1)+
  labs(x="Years",
       y="N° of sales",
       title="Sales line chart from 2010 to 2014")



#generates array containing the n° sales per month and than plots the result
sales_history_month=c()
k=1
for (i in min(year):max(year)){
  for (j in min(month):max(month)){
    sales_history_month[k]=sum(sales[year==i & month==j])
    k=k+1
  }
}

ggplot()+
  geom_line(aes(x=c(1:60),y=sales_history_month),lwd=1)+
  labs(x="Months",
       y="N° of sales",
       title="Line chart from 2010 to 2014")+
  scale_x_continuous(breaks=seq(0,60,12))



#comparison of line charts for each city
dati_Beaumont=filter(dati, city=="Beaumont") 
dati_BCS=filter(dati, city=="Bryan-College Station")
dati_Tyler=filter(dati, city=="Tyler") 
dati_Wichita_Falls=filter(dati, city=="Wichita Falls")

ggplot()+
  geom_line(aes(x=c(1:60),y=dati_Beaumont$sales,col="Beaumont"),lwd=1)+
  geom_line(aes(x=c(1:60),y=dati_BCS$sales,col="Bryan-College Station"),lwd=1)+
  geom_line(aes(x=c(1:60),y=dati_Tyler$sales,col="Tyler"),lwd=1)+
  geom_line(aes(x=c(1:60),y=dati_Wichita_Falls$sales,col="Wichita Falls"),lwd=1)+
  labs(x="Months",
       y="N° of sales",
       title="Line chart from 2010 to 2014 for each city")+
  scale_x_continuous(breaks=seq(0,60,12))+
  theme(legend.position="bottom")

#changes colors
ggplot()+
  geom_line(aes(x=c(1:60),y=dati_Beaumont$sales,col="Beaumont"),lwd=1)+
  geom_line(aes(x=c(1:60),y=dati_BCS$sales,col="Bryan-College Station"),lwd=1)+
  geom_line(aes(x=c(1:60),y=dati_Tyler$sales,col="Tyler"),lwd=1)+
  geom_line(aes(x=c(1:60),y=dati_Wichita_Falls$sales,col="Wichita Falls"),lwd=1)+
  labs(x="Months",
       y="N° of sales",
       title="Line chart from 2010 to 2014 for each city")+
  scale_color_manual(name="",
                     breaks=c("Beaumont","Bryan-College Station","Tyler","Wichita Falls"),
                     values=c("red","green2","lightblue","orange"),
                     labels=c("Beaumont","Bryan-College Station","Tyler","Wichita Falls"))+
  scale_x_continuous(breaks=seq(0,60,12))+
  theme(legend.position="bottom")

#different method, but equal results
sales_history_month_Beaumont=c()
sales_history_month_BCS=c()
sales_history_month_Tyler=c()
sales_history_month_Wichita_Falls=c()
k=1
for (i in min(year):max(year)){
  for (j in min(month):max(month)){
    sales_history_month_Beaumont[k]=sum(sales[city=="Beaumont" & year==i & month==j])
    sales_history_month_BCS[k]=sum(sales[city=="Bryan-College Station" & year==i & month==j])
    sales_history_month_Tyler[k]=sum(sales[city=="Tyler" & year==i & month==j])
    sales_history_month_Wichita_Falls[k]=sum(sales[city=="Wichita Falls" & year==i & month==j])
    k=k+1
  }
}

ggplot()+
  geom_line(aes(x=c(1:60),y=sales_history_month_Beaumont,col="Beaumont"),lwd=1)+
  geom_line(aes(x=c(1:60),y=sales_history_month_BCS,col="Bryan-College Station"),lwd=1)+
  geom_line(aes(x=c(1:60),y=sales_history_month_Tyler,col="Tyler"),lwd=1)+
  geom_line(aes(x=c(1:60),y=sales_history_month_Wichita_Falls,col="Wichita Falls"),lwd=1)+
  labs(x="Months",
       y="N° of sales",
       title="Line chart from 2010 to 2014 for each city")+
  scale_color_manual(name="",
                     breaks=c("Beaumont","Bryan-College Station","Tyler","Wichita Falls"),
                     values=c("red","green2","lightblue","orange"),
                     labels=c("Beaumont","Bryan-College Station","Tyler","Wichita Falls"))+
  scale_x_continuous(breaks=seq(0,60,12))+
  theme(legend.position="bottom")

```
